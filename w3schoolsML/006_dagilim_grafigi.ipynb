### DAĞILIM GRAFİĞİ

Dağılım grafiği, veri kümesindeki her değerin bir nokta ile temsil edildiği bir diyagramdır.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
ÖRNEK 1: Matplotlib modülünün dağılım grafikleri çizmek için bir yöntemi vardır, aynı uzunlukta iki diziye ihtiyaç duyar, 
biri x ekseni değerleri için ve diğeri y ekseni değerleri için:
"""
x = [5,7,8,7,2,17,2,9,4,11,12,9,6] # Dizi x , her arabanın yaşını temsil eder.
y = [99,86,87,88,111,86,103,87,94,78,77,85,86] # Dizi y , her arabanın hızını temsil eder.

plt.scatter(x, y) # scatter() yöntemi, dağılım grafiği çizmek için kullanılır.
plt.show()
```

### RASTGELE VERİ DAĞILIMLARI

```python
x = np.random.normal(5.0, 1.0, 1000)  
y = np.random.normal(10.0, 2.0, 1000) 

plt.scatter(x, y)
plt.show()
```
