```python
import numpy as np
import matplotlib.pyplot as plt
```

### SORU 1 : Büyük Veri Kümelerini Nasıl Elde Edebiliriz?

```python
# 0 ile 5 arasında 250 adet ondalıklı sayı üret.

x=np.random.uniform(0,5,250)
print(x)
```

```python
# Veriyi histogram olarak gösterelim.

plt.hist(x, 5) # 5 tane çubuk oluştur.
plt.show()
```

```python
# ÖRNEK 2 : 100000 rasgele sayıdan oluşan bir dizi oluşturun ve bunları 100 çubuklu bir histogram kullanarak görüntüleyin:

y = np.random.uniform(0.0, 5.0, 100000)
plt.hist(y, 100)
plt.show()
```
