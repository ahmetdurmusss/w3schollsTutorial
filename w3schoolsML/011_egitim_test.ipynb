```python
# Veri setimiz bir mağazadaki 100 müşteriyi ve onların alışveriş alışkanlıklarını göstermektedir.
import numpy
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
numpy.random.seed(2)

x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

plt.scatter(x, y)
plt.show()
```

```python
#Eğitim seti , orijinal verilerin %80'inin rastgele bir seçimi olmalıdır.Test seti kalan %20 olmalıdır.
train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

```

```python
plt.title("Eğitim Verileri")
plt.scatter(train_x, train_y)
plt.show()
```

```python
plt.title("Test Verileri")
plt.scatter(test_x, test_y)
plt.show()
```

```python
mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

myline = numpy.linspace(0, 6, 100)

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
#R Kare Değeri
r2_train= r2_score(train_y, mymodel(train_x))
r2_test = r2_score(test_y, mymodel(test_x))
print("R^2 TRAİN:" ,r2_train)
print("R^2 TEST:" ,r2_test)
```
