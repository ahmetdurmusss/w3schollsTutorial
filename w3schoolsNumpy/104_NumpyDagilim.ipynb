## Normal (GAUSS) Dağılımı

Normal Dağılım en önemli dağılımlardan biridir.Alman matematikçi Carl Friedrich Gauss'un adıyla Gauss Dağılımı olarak da adlandırılır. Birçok olayın olasılık dağılımına uyar, örneğin; IQ Puanı, Kalp Atışı vb.

Normal Veri Dağılımını elde etmek için random.normal() yöntemi kullanılır.

Üç parametresi vardır:

loc - Çan sesinin tepe noktasının bulunduğu yer.

scale - Grafik dağılımının ne kadar düz olması gerektiği.

size - Döndürülen dizinin şekli.


```python
# ÖRNEK 1 : 2x3 boyutunda rastgele normal dağılım oluşturalım.

from numpy import random
x = random.normal(size=(2, 3))
print(x)
```

```python
# ÖRNEK 2 : Ortalaması 1 ve standart sapması 2 olan 2x3 boyutunda rastgele normal dağılım oluşturalım.
from numpy import random

x = random.normal(loc=1, scale=2, size=(2, 3))

print(x)
```

```python
# NORMAL DAĞILIMIN GÖRSELLEŞTİRİLMESİ 

# ÖRNEK 3 : 

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(size=1000), hist=False)

plt.legend(['Normal Dağılım'])
plt.show()
```

## Binom Dağılım

Binom Dağılımları Ayrık Dağılımlardır.

İkili senaryoların sonucunu açıklar, örneğin paranın havaya atılması sonucu yazı ya da tura gelmesi.

Üç parametresi vardır:

n- Deneme sayısı.

p- Her bir denemenin gerçekleşme olasılığı (örneğin yazı tura atıldığında her biri 0,5).

size- Döndürülen dizinin şekli.

##### NOT : Ayrık Dağılım: Dağılım, ayrı bir olay kümesinde tanımlanır. Örneğin, yazı tura atışının sonucu yalnızca yazı veya tura olabileceğinden ayrıktır; ancak insanların boyları 170, 170.1, 170.11 vb. olabileceğinden süreklidir.

```python
# ÖRNEK 4 : 10 denemde yazı tura atılarak 10 veri noktası oluşturalım.

from numpy import random
x = random.binomial(n=10, p=0.5, size=10)
print(x)
```

```python
# BİNOM DAĞILIMININ GÖRSELLEŞTİRİLMESİ
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True, kde=False)

plt.legend(['Binom Dağılım'])
plt.show()
```

### BİNOM VE NORMAL DAĞILIM ARASINDAKİ FARK
Temel fark, normal dağılımın sürekli, binom dağılımının ise kesikli olmasıdır; ancak yeterli sayıda veri noktası varsa, belirli konum ve ölçekte normal dağılıma oldukça benzer olacaktır

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(loc=50, scale=5, size=1000), hist=False, label='normal')
sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=False, label='binomial')

plt.legend()
plt.show()
```

## Poisson Dağılımı

Poisson Dağılımı ayrık bir dağılımdır .
Belirli bir zaman diliminde bir olayın kaç kez gerçekleşebileceğini tahmin eder. Örneğin; birisi günde iki kez yemek yiyorsa, üç kez yeme olasılığı nedir?

İki parametresi vardır:

lam- Yukarıdaki problem için örneğin 2 gibi bilinen oluşum sayısı veya oranı.

size- Döndürülen dizinin şekli.

```python
# ÖRNEK 5 : 2. olay için rastgele 1x10'luk bir dağılım oluşturalım.
from numpy import random

x = random.poisson(lam=2, size=10)

print(x)
```

#### Poisson Dağılımının Görselleştirilmesi

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.poisson(lam=2, size=1000), kde=False)

plt.legend(['Poisson Dağılım'])
plt.show()
```

#### Normal ve Poisson Dağılımı Arasındaki Fark
Normal dağılım süreklidir, poisson ise ayrıktır.Ancak yeterince büyük bir poisson dağılımı için binom dağılımına benzer şekilde, belirli standart sapma ve ortalama değerleri ile normal dağılıma benzer hale geleceğini görebiliriz.

```python
# ÖRNEK 6 :
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(loc=50, scale=7, size=1000), hist=False, label='normal') 
sns.distplot(random.poisson(lam=50, size=1000), hist=False, label='poisson')

plt.legend()
plt.show()
```

#### Binom ve Poisson Dağılımı Arasındaki Fark

Binom dağılımının yalnızca iki olası sonucu vardır, oysa poisson dağılımının sınırsız olası sonuçları olabilir.Ancak çok büyük n ve sıfıra yakın p için binom dağılımı poisson dağılımıyla hemen hemen aynıdır, öyle ki n*p neredeyse lam'e eşittir.


```python
# ÖRNEK 7 :
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.binomial(n=1000, p=0.01, size=1000), hist=False, label='binomial')
sns.distplot(random.poisson(lam=10, size=1000), hist=False, label='poisson')

plt.legend()
plt.show()
```

## UNIFORM DAĞILIM
Her olayın gerçekleşme şansının eşit olduğu olasılığı tanımlamak için kullanılır.

Örneğin, Rastgele sayıların üretilmesi.

Üç parametresi vardır:

a- alt sınır - varsayılan 0 .0.

b- üst sınır - varsayılan 1.0.

size- Döndürülen dizinin şekli.

```python
# ÖRNEK 8 : 2x3'lük uniform dağılımlı bir örnek oluşturalım.
from numpy import random
x = random.uniform(size=(2, 3))
print(x)
```

```python
# UNIFORM DAĞILIMININ GÖRSELLEŞTİRİLMESİ

# ÖRNEK 9 :
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.uniform(size=1000), hist=False)

plt.legend(['Uniform Dağılım'])
plt.show()
```

## LOJİSTİK DAĞILIM

Büyümeyi tanımlamak için Lojistik Dağıtım kullanılır.
Makine öğrenmesinde, lojistik regresyonda, sinir ağlarında vb. yaygın olarak kullanılır.

Üç parametresi vardır:

loc- ortalama, zirvenin olduğu yer. Varsayılan 0.

scale- standart sapma, dağılımın düzlüğü. Varsayılan 1.

size- Döndürülen dizinin şekli.

```python
# ÖRNEK 10 : Ortalaması 1 ve standart sapması 2,0 olan bir lojistik dağılımdan 2x3'lük örnek oluşturalım.

from numpy import random

x = random.logistic(loc=1, scale=2, size=(2, 3))

print(x)
```

```python
# LOJİSTİK DAĞILIMININ GÖRSELLEŞTİRİLMESİ

# ÖRNEK 11 :

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.logistic(size=1000), hist=False)

plt.legend(['Logistic Dağılım'])
plt.show()
```

##### Lojistik ve Normal Dağılım Arasındaki Fark
Her iki dağılım da neredeyse aynıdır, ancak lojistik dağılımın grafik altında daha fazla alanı vardır, bu da ortalamadan daha uzakta bir olayın meydana gelme olasılığının daha yüksek olduğu anlamına gelir.

Ölçek değerinin (standart sapma) daha yüksek olması durumunda, tepe noktası dışında normal ve lojistik dağılımlar hemen hemen aynıdır.

```python
# ÖRNEK 12 : 

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(scale=2, size=1000), hist=False, label='normal')
sns.distplot(random.logistic(size=1000), hist=False, label='logistic')

plt.legend()
plt.show()
```

## ÇOK TERİMLİ DAĞILIM

Çok terimli dağılım, iki terimli dağılımın genelleştirilmiş halidir.

Binom senaryolarında senaryoların ikisinden yalnızca biri olması gerekirken, çok adlandırmalı senaryoların sonuçlarını açıklar. Örneğin, bir popülasyonun kan grubu, zar atma sonucu.

Üç parametresi vardır:

n- olası sonuçların sayısı (örneğin zar atıldığında 6).

pvals- Sonuçların olasılık listesi (örneğin zar atışı için [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]).

size- Döndürülen dizinin şekli.

```python
# ÖRNEK 13 : Zar atma örneği
from numpy import random

x = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

print(x)
```

## ÜSTEL DAĞILIM
Üstel dağılım, bir sonraki olaya (başarı/başarısızlık vb.) kadar geçen süreyi tanımlamak için kullanılır.

İki parametresi vardır:

scale- oranın tersi (bkz. Poisson dağılımındaki lam) varsayılan olarak 1.0'dır.

size- Döndürülen dizinin şekli.

```python
# ÖRNEK 14 : 2x3 boyutunda 2.0 ölçekli üstel dağılım için bir örnek çizin:

from numpy import random

x = random.exponential(scale=2, size=(2, 3))

print(x)
```

```python
# ÜSTEL DAĞILIMININ GÖRSELLEŞTİRİLMESİ

# ÖRNEK 15 :
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.exponential(size=1000), hist=False)

plt.legend(['Üstel Dağılım'])
plt.show()
```
