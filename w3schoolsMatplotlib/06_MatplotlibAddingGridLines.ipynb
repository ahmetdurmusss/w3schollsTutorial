## PYTHON GRAFİKLERİ IZGARALAMA 
Pyplot ile çizime ızgata çizgileri ekleme " grid() " fonksiyonunu kullanabilirsiniz.

```python
# ÖRNEK 1 :Çizgi grafiği çizimi ve ızgara ekleme

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y) # çizgi grafiği çizimi
plt.grid() # ızgara ekleme
plt.show()
```

```python
# ÖRNEK 2 : Çizgi grafiği çizimi ve sadece x eksenine ızgaraları gösterme

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'x')#sadece x eksenine göre ızgara olacak
```

```python
# ÖRNEK 3 : Çizgi grafiği çizimi ve sadece y eksenine ızgaraları göster
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'y')

plt.show()
```

#### Izgara Çizgiler İçin Ayarlama Yapmak

Izgara çizgi özelliklerini şu şekilde ayarlayabilirsiniz: 
grid(color = ' color ', linestyle = ' linestyle ', linewidth = number ).

```python
# ÖRNEK 4 : Çizgi grafiği çizimi ve ızgara rengini ve stilini değiştirme.

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.grid(color = 'hotpink', linestyle = ':', linewidth = 0.5) # ızgara rengini ve stilini değiştirme
plt.show()
```
