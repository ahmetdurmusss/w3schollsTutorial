### PASTA GRAFİĞİ OLUŞTURMAK
Pyplot ile pasta grafikler çizmek için "pie()" işlevi kullanabilirsiniz

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ÖRNEK : BASİT PASTA GRAFİĞİ OLUŞTURMA
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

### PASTA GRAFİĞİNDE PARÇA ETİKETLEME 
"label" parametresi ile pasta grafiğine etiketler ekleyin.

"label" parametresi, her dilim için bir etiket içeren bir dizi olmalıdır.



```python
# ÖRNEK : PASTA GRAFİĞİNE ETİKET EKLEME

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

### BAŞLANGIÇ AÇISINI AYARLAMA
Bahsedildiği gibi, varsayılan başlangıç ​​açısı x eksenindedir, ancak başlangıç ​​açısını bir parametre belirleyerek değiştirebilirsiniz: "startangle"
"startanglederece" parametresi cinsinden bir açıyla tanımlanır, varsayılan açı 0'dır.

```python
# ÖRNEK: İLK PARÇAYI 90 DERECEDE BAŞLATMAK

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

### GRAFİKTEN PARÇA ÇIKARMAK
Grafikten bir parça çıkarmak için "explode" parametresini kullanın. "explode" parametresi, her dilim için bir değer içeren bir dizi olmalıdır.


```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

### GÖLGELİ PASTA GRAFİĞİ
"shadows= True" şeklinde ayarlayarak pasta grafiğe bir gölge ekleyin.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

### PASTA GRAFİĞİ RENKLENDİRME
"colors=" parametresi ile her takozun rengini ayarlayabilirsiniz.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

### YAZIT EKLEME 
her dilime bir açıklama listesi eklemek için "legend()" işlevi kullanın.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

```python
# ÖRNEK: PASTA GRAFİĞİNE BAŞLIKLI YAZIT EKLEMEK

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```
