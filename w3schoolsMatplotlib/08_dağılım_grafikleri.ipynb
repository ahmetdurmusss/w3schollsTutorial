### DAĞILIM GRAFİKLERİ OLUŞTURMA 
Pyplot ile, bir dağılım grafiği çizmek için "scatter()" işlevi kullanabilirsiniz.
scatter() İşlevi, her gözlem için bir nokta çizer. Biri x eksenindeki değerler için ve diğeri y eksenindeki değerler için aynı uzunlukta iki diziye ihtiyaç duyar.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ÖRNEK : BASİT BİR DAĞILIM GRAFİĞİ

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

### GRAFİKLERİ KARŞILAŞTIRMA 

```python
# BİRİNCİ GÜN , 13 ARABA HIZI VE YAŞI:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#İKİNCİ GÜN , 15 ARABA HIZI VE YAŞI:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

### NOKTALARIN RENGİNİ DEĞİŞTİRME.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#88c999')

plt.show()
```

### HER NOKTATI RENKLENDİRME

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

```python
# RENK HARİTASIYLA RENKLİ NOKTALARIN GRAFİĞİNİ OLUŞTURMAK.

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

### NOKTA BOYUTUNU DEĞİŞTİRME
's' argüman ile noktaların boyutunu değiştirebilirsiniz.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

### RENGİN ŞEFFAFLIĞINI DEĞİŞTİRME
alpha argüman ile noktaların şeffaflığını ayarlayabilirsiniz.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.7)

plt.show()
```

```python
# KONU ÖRNEĞİ:

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
