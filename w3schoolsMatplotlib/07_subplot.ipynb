### BİRDEN FAZLA GRAFİĞİ AYNI EKRANDA GÖSTERMEK
subplot(), birden fazla grafik çizmek için kullanılır. subplot() fonksiyonu 3 parametre alır. Biri satır sayısı, biri sütun sayısı, biri de hangi grafikte olduğumuzu belirten index değeridir.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ÖRNEK : YAN YANA 2 GRAFİK ÇİZİMİ
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
# ÖRNEK : ÜST ÜSTE 2 GRAFİK ÇİZİMİ
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

```python
# ÖRNEK : 6 TANE GRAFİK ÇİZİMİ

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```

### BAŞLIK EKLEME
title() fonksiyonu ile grafiklere başlık ekleyebiliriz.

```python
# ÖRNEK : YAN YANA ÇİZİLEN GRAFİKLERE BAŞLIK EKLEME.
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SATIŞ")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("GELİR")

plt.show()
```

### !!!!!
YAN YANA ÇİZİLEN İKİ GRAFİĞE ORTAK BAŞLIK EKLEME.

subtitle() fonksiyonu ile grafiklere ortak başlık ekleyebiliriz.

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SATIŞ")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("GELİR")

plt.suptitle("DÜKKAN")
plt.show()
```
