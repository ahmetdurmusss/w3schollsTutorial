## PYTHON BAR GRAFİĞİ OLUŞTURMA
Pyplot ile " bar() " fonksiyonu kullanarak çubuk grafikler çizebilirsiniz.

```python
# ÖRNEK 1 : Çubuk Grafiği

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y) # Çubuk grafiği oluşturuluyor.
plt.show()
```

```python
# ÖRNEK 2 : Çubuk Grafiği

import matplotlib.pyplot as plt
import numpy as np

x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)
plt.show()
```

```python
# BAR GRAFİĞİNDE YATAY ÇUBUKLAR : Çubukların dikey yerine yatay olarak görüntülenmesini istiyorsanız şu " barh() " fonksiyonu kullanılır.
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

#### Çubuk Grafiğini Renklendirme
" bar() " ve " barh() " çubukların rengini ayarlamak için " color " anahtar sözcük argümanını alır :

```python
# ÖRNEK 4 : Kırmızı renkli çubuk grafiği
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red")
plt.show()
```

```python
# ÖRNEK 5 : Çubuk grafiğinde hexadecimal renk değerleri
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "#4CAF50") # hexadecimal renk değerleri
plt.show()
```

#### Grafikteki Çubukların Gnişliği
Çubukların genişliğini ayarlamak için bar() fonksiyonu " width " anahtar sözcük argümanını alır.

```python
# ÖRNEK 6 : Çubuk grafiğinde genişlik ayarı.

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1) # Çubukların genişliği
plt.show()
```

#### Bar Yüksekliği
Çubukların yüksekliğini ayarlamak için " barh() " fonksiyonu " height " anahtar sözcük argümanını alır.

**Örnek
4 tane çok ince çubuk çizin:**

```python
# ÖRNEK 7 : Çubuk grafiğinde yükseklik ayarı.
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```

## PYTHON MATPLOTLİB HİSTOGRAM GRAFİĞİ
Histogram, frekans dağılımlarını gösteren bir grafiktir. Belirli bir aralıktaki gözlem sayısını gösteren bir grafiktir.
Matplotlib'de " hist() " histogram oluşturmak için bu fonksiyonu kullanıyoruz. " hist() " fonksiyonu, bir histogram oluşturmak için sayılardan oluşan bir dizi kullanacak ve dizi, fonksiyona argüman olarak gönderilecek.

```python
# ÖRNEK 8 : 20 tane rastgele sayı üretip histogramını çizdiriniz.
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```
