## Matplotlib İle Grafik Çizimi

x ve y noktalarının çizilmesi
1) " plot() " fonksiyonu bir diyagramda noktaları (işaretleyicileri) çizmek için kullanılır.

2) " plot() " fonksiyonu varsayılan olarak  noktadan noktaya bir çizgi çizer.Fonksiyon, diyagramdaki noktaları belirtmek için parametreler alır.

Parametre 1, x eksenindeki noktaları içeren bir dizidir .

Parametre 2, y eksenindeki noktaları içeren bir dizidir .

Örnek
Bir diyagramda (1, 3) pozisyonundan (8, 10) pozisyonuna bir çizgi çizin:

```python
# ÖRNEK 1: İki nokta arasında çizgi çizme. (1,3) ve (8,10) noktaları arasında çizgi çizme.

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8]) # x eksenindeki noktalar
ypoints = np.array([3, 10]) # y eksenindeki noktalar 

plt.plot(xpoints, ypoints) # Çizgi çizme
plt.show()
```

#### Noktasal Çizim
Sadece işaretleyicileri çizmek için, 'halkalar' anlamına gelen 'o' kısayol dizesi gösterim parametresini kullanabilirsiniz .

```python
# ÖRNEK 2 : (1,3) ve (8,10) noktalarına işaretçi koyma.

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o') # 'o' işaretçi koyma
plt.show()
```

#### Çoklu Çizim
İstediğiniz kadar nokta çizebilirsiniz, sadece her iki eksende aynı sayıda nokta olduğundan emin olun.

```python
# ÖRNEK 3 : (1,3), (2,8), (6,1) ve (8,10) noktaları arasında çizgi çizme.

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8]) # x eksenindeki noktalar
ypoints = np.array([3, 8, 1, 10]) # y eksenindeki noktalar

plt.plot(xpoints, ypoints) # Çizgi çizme
plt.show() # Çizgiyi gösterme
```

Default/Varsayılan X-Points/Noktaları

Eğer x eksenindeki noktaları belirtmezsek, y noktalarının uzunluğuna bağlı olarak varsayılan 0, 1, 2, 3 vb. değerleri alacaklardır.

Örnek
X noktaları olmadan çizim:

```python
# NOT : Eğer yalnızca y noktalarını verirseniz, x noktaları kendiliğinden gelir. Bu durumda x noktaları 0,1,...,n  olur.

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)#buradada yalnızca y noktasını almış parametre olarak x kendiliğinden gelmiş mesela
plt.show()
```

#### " marker " 
Her noktayı belirtilen bir işaretleyiciyle vurgulamak için " marker " anahtar sözcüğünü kullanabilirsiniz. Girdiğiniz değer string formatında yazılmalı.

İşaretçi tipi
1) 'o'	------> Circle	
2) '*'	------> Star	
3) '.'	------> Point	
4) ','	------> Pixel

Çizgi tipi değerleri
1) ' -  '	------> Solid line	
2) ' :  '	------> Dotted line	
3) ' -- '	------> Dashed line	
4) ' -. '	------> Dashed/dotted line


Renkler değerleri
1) 'r' ------> Red	
2) 'g' ------> Green	
3) 'b' ------> Blue	
4) 'c' ------> Cyan	
5) 'm' ------> Magenta	
6) 'y' ------> Yellow	
7) 'k' ------> Black	
8) 'w' ------> White

```python
# ÖRNEK 4 : (1,3), (2,8), (6,1) ve (8,10) noktalarına işaretçi koyma.

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10]) # y eksenindeki noktalar

plt.plot(ypoints, marker = 'o') # 'o' işaretçi koyma
plt.show()
```

```python
# ÖRNEK 5 : (1,3), (2,8), (6,1) ve (8,10) noktalarına * işaretçi koyma.

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = '*') # '*' işaretçi koyma
plt.show()
```

```python
# ÖRNEK 6 : (1,3), (2,8), (6,1) ve (8,10) noktalarına işaretçi koyma ve kırmızı renk verme.
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r') # 'o' işaretçi koyma ve kırmızı renk verme. (marker:line:color)
plt.show()
```

#### İşaretçi Boyutu (markersize)
İşaretçilerin boyutunu ayarlamak için " markersize " veya daha kısa versiyonunu olan " ms " anahar kelimelerimi kullanabilirsiniz.

```python
# ÖRNEK 7 : (1,3), (2,8), (6,1) ve (8,10) noktalarına işaretçi koyma 

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20) # 'o' işaret , ms = 20 işaretçi boyutu
plt.show()
```

#### İşaretçi Rengi (markeredgecolor)

İşaretçilerin boyutunu ayarlamak için " markeredgecolor " veya daha kısa versiyonunu olan " mec " anahar kelimelerimi kullanabilirsiniz.

```python
# ÖRNEK 8 : (1,3), (2,8), (6,1) ve (8,10) noktalarına işaretçi koyma ve işaretçi rengini yeşil yapma.

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'g') # İşaretçinin kenar rengini yeşil yapma
plt.show()
```

```python
# NOT : İşaretçinin kenar yerine iç rengini değiştirmek için mfc parametresi kullanılır.

# ÖRNEK 9 : (1,3), (2,8), (6,1) ve (8,10) noktalarına işaretçi koyma ve işaretçi rengini kırmızı yapma.

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r',mec='y') # İşaretçinin iç rengini kırmızı yapma
plt.show()
```
