### HİSTOGRAM
Histogram, frekans dağılımlarını gösteren bir grafiktir .

Belirli bir aralıktaki gözlem sayısını gösteren bir grafiktir.

### HİSTOGRAM OLUŞTURMA
Matplotlib'de, histogramlar oluşturmak için "hist()" işlevi kullanırız .

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ÖRNEK : NumPy ile NOORMAL BİR VERİ DAĞILIMI İLE HİSTOGRAM ÇİZDİRME.

x = np.random.normal(170, 10, 250)


plt.hist(x)
plt.show()
print(x)
```
