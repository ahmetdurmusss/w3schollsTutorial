### ÇİZGİ STİLİ 
Çizilen çizginin stilini değiştirmek için "linestyle" veya daha kısa olan "ls" anahtar kelimesini kullanabilirsiniz.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ÖRNEK : Noktalı bir çizgi kullanın.

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
# ÖRNEK : Kesikli bir çizgi kullanın.

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls = ':')
plt.show()
```

### ÇİZGİ RENGİ
Satırın rengini ayarlamak için anahtar kelime olan "color" bağımsız değişkenini veya daha kısa olan "c" kullanabilirsiniz.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = '#4CAF50')
plt.show()
```

### ÇİZGİ GENİŞLİĞİ
Satırın genişliğini değiştirmek için anahtar kelime "linewidth" veya daha kısa olan "lw" kullanabilirsiniz.

```python
# ÖRNEK : 20,5 punto geniş bir çizgi 

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

### ÇOKLU SATIR 
Sadece "plt.plot()" fonksiyon ekleyerek istediğiniz kadar çizgi çizebilirsiniz.

```python
# ÖRNEK:
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1 , lw='2')
plt.plot(y2 , lw= '3.5')

plt.show()
```

```python
# ÖRNEK: Her iki çizgi için de x ve y noktalarını belirterek iki çizgi çizin.

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```
