### GRAFİK İÇİN ETİKETLER OLUŞTURMA 
Pyplot ile, x ve y ekseni için bir etiket ayarlamak üzere xlabel() ve ylabel() işlevlerini kullanabilirsiniz .

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ÖRNEK : x ve y eksenine etiket ekleyin.
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Ortalama Nabız")
plt.ylabel("Kalori Yakma")

plt.show()
```

### Bir Konu İçin Başlık Oluşturun
Pyplot ile arsa için bir başlık ayarlamak üzere "title()" işlevi kullanabilirsiniz .

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Spor Saati Verileri")
plt.xlabel("Ortalama Nabız")
plt.ylabel("Kalori Yakma")

plt.show()
```

### Başlık ve Etiketler için Yazı Tipi Özelliklerini Ayarlama
Başlık ve etiketlerin yazı tipi özelliklerini ayarlamak için "xlabel()","ylabel()","title()" ve "fontdict()" içindeki parametreyi kullanabilirsiniz .

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Spor Saati Verileri", fontdict = font1)
plt.xlabel("Ortalama Nabız", fontdict = font2)
plt.ylabel("Kalori Yakma", fontdict = font2)

plt.plot(x, y)
plt.show()
```

### Başlığı Konumlandırma 
Başlığı konumlandırmak için "loc()" parametresini kullanabilirsiniz.

Yasal değerler şunlardır: 'sol (left)', 'sağ (right)' ve 'merkez (center)'. Varsayılan değer 'merkez (center)' dir.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Spor Saati Verileri" , loc = "left")
plt.xlabel("Ortalama Nabız")
plt.ylabel("Kalori Yakma")

plt.plot(x, y)
plt.show()
```
