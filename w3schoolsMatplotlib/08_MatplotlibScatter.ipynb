## PYTHON MATPLOTLİB DAĞILIM GRAFİĞİ OLUŞTURMA

Pyplot ile dağılım grafiği çizmek için " scatter() " foksiyonunu kullanırız.Fonksiyon " scatter() " her gözlem için bir nokta çizer. Aynı uzunlukta iki diziye ihtiyaç duyar, biri x eksenindeki değerler için, diğeri y eksenindeki değerler için.

```python
# ÖRNEK 1 : Basit bir dağıım grafiği çizimi

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6]) # x ve y eksenlerine ait verileri tanımlıyoruz.
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86]) # x ve y eksenlerine ait verileri tanımlıyoruz.

plt.scatter(x, y) # scatter() fonksiyonu ile dağılım grafiği çizimi yapılır.
plt.show()
```

```python
# AYNI TUVALDE İKİ ADET DAĞILIM GRAFİĞİ ÇİZİMİ : 
# İki farklı gözlem birimine ait dağılım grafiğini aynı tuvalde çizmek için aşağıdaki örneği inceleyebilirsiniz.

# ÖRNEK 2 : İki adet dağılım grafiğini aynı tuvalde çizimi
import matplotlib.pyplot as plt
import numpy as np

# Grafik 1, 13 gözlem birimi
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)
plt.show()
```

#### Dağılım Grafiği Renklendirme
Her dağılım grafiği için kendi renginizi " color " veya " c " argümanını kullanarak ayarlayabilirsiniz.

```python
# ÖRNEK 3 : Renkli dağılım grafiği çizimi
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink') # Renk değeri belirtilir. 

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#88c999') # Renk değeri belirtilir.

plt.show()
```

```python
# NOT : Dağılım grafiğindeki her noktayı farklı renkte kullanabilisiniz.

# ÖRNEK 4 : Renkli dağılım grafiği çizimi
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

#### Renk Haritası Oluşturma 
Renk haritası, her rengin 0 ile 100 arasında değişen bir değere sahip olduğu bir renk listesi gibidir.

Renk haritası oluşturmak için matplotlib kütüphanesindeki colormap() fonksiyonu kullanılır. Bu fonksiyon ile renk haritası oluşturulur. Bu fonksiyonun aldığı parametreler, cmap ve c olabilir. cmap parametresi ile renk haritası belirlenirken, c parametresi ile renklerin değerleri belirlenir.

Ayrıca, dağılım grafiğindeki her nokta için bir değer olmak üzere 0'dan 100'e kadar değerler içeren bir dizi oluşturmanız gerekir.

```python
# ÖRNEK 5 : Renkli dağılım grafiği çizimi ve renk skalası

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis') # Renk skalası belirtilir. viridis, plasma, inferno, magma, cividis gibi renk skalaları kullanılabilir.

plt.colorbar()

plt.show()
```

```python
# ÖRNEK 6 : Renkli dağılım grafiği çizimi ve renk skalası

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='cividis') # Renk skalası belirtilir. viridis, plasma, inferno, magma, cividis gibi renk skalaları kullanılabilir.

plt.colorbar()

plt.show()
```

#### Dağılım Grafiklerinde Boyut Belirleme
Noktaların boyutunu " s " argümanı ile değiştirebilirsiniz. Tıpkı renklerde olduğu gibi, boyutlar dizisinin x ve y eksenlerindeki dizilerle aynı uzunlukta olduğundan emin olun.

```python
# ÖRNEK 7 : Renkli dağılım grafiği nokta boyutu belirleme

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes) # s : nokta boyutu belirtilir.

plt.show()
```

```python
# NOKTALARIN ŞEFFAFLIĞINI AYARLAMA : Dağılım grafinde yer alan noktaların şeffaflığını ayarlamak için alpha parametresi kullanılır.

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.2) # alpha : noktaların şeffaflığı belirtilir.
plt.show()
```

#### Renk Boyutunu ve Alfa'yı Birleştir
Farklı boyutlardaki noktalarla bir renk haritasını birleştirebilirsiniz. Bu, noktalar şeffafsa en iyi şekilde görselleştirilir.

```python
# ÖRNEK 8 : Renkli ve şeffaf dağılım grafiği çizimi

import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral') 
plt.show()
```
