### BİR GRAFİĞE IZGARA ÇİZGİLERİ EKLEME
Pyplot ile , grafiğe ızgara çizgileir eklemek için grid() fonksiyonu kullanılır.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ÖRNEK : Grafiğe ızgara çizgileri ekleyin.
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.grid()
plt.show()
```

### !!!!
ÇİGİLERİN HANGİ YÖNDE OLANLARININ GÖZÜKMESİ GEREKTİĞİNİ axis() FONKSİYONU İLE BELİRLEYEBİLİRİZ.

" plt.grid(axis = 'x' veya 'y' )"

```python
# ÖRNEK : YANLIZCA DİKEY ÇİZGİLER (-x ekseni).
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'x')

plt.show()
```

```python
# ÖRNEK : YANLIZCA YATAY ÇİZGİLER (-y ekseni).
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'y')

plt.show()
```

### Izgara için Çizgi Özelliklerini Ayarlama
Izgaranın çizgi özelliklerini de şu şekilde ayarlayabilirsiniz: grid(renk = ' color ', cizgi stili = ' linestyle ', kalınlık = number ).

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(color = 'green', linestyle = '--', linewidth = 1.5)

plt.show()
```
