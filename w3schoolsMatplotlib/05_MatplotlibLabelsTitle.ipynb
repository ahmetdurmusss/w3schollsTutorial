## PYTHON GRAFİK ETİKETLEME - İSİMLENDİRME
Pyplot ile x ve y eksenleri için bir etiket belirlemek amacıyla xlabel()ve fonksiyonlarını kullanabilirsiniz . ylabel()

**Örnek
X ve y eksenlerine etiketler ekleyin:**

```python
# ÖRNEK 1: Bir grafik çizelim. x eksenine ortalama nabız değerlerini, y eksenine ise kalori yakımını yerleştirelim.

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125]) # Ortalama nabız değerleri
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330]) # Kalori yakımı

plt.plot(x, y)

plt.xlabel(" ORTALAMA NABIZ ")
plt.ylabel(" KALORİ YAKIMI ")

plt.show()
```

```python
# GRAFİK ÜZERİNDE BAŞLIK EKLEME : title() fonksiyonu kullanılır.

# ÖRNEK 2: Bir grafik çizelim. x eksenine ortalama nabız değerlerini, y eksenine ise kalori yakımını yerleştirelim. Ayrıca grafik üzerine başlık ekleyelim.

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Spor Saati Verileri") # Grafik başlığı
plt.xlabel("Ortalama Nabız") # x ekseninin adı
plt.ylabel("Kalori Yakımı") # y ekseninin adı
plt.show()
```

#### Başlık ve Etiketler İçin Yazı Tipi Seçimi 

Python da grafiklerimiz için başlık ve eksen isimlendirme işlemlerini " fontdict " parametresini kullanarak yapabiliriz.

**Örnek
Başlık ve etiketler için yazı tipi özelliklerini ayarlayın:**

```python
# ÖRNEK 3: Grafik üzerinde başlık, x ekseninin adı ve y ekseninin adını farklı font ve renklerde yazdıralım.

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'hotpink','size':20} # Başlık fontu {family, color, size}
font2 = {'family':'serif','color':'darkred','size':15} # Eksen adları fontu 

plt.title("Sports Watch Data", fontdict = font1) # Grafik başlığı
plt.xlabel("Average Pulse", fontdict = font2)  # x ekseninin adı 
plt.ylabel("Calorie Burnage", fontdict = font2) # y ekseninin adı

plt.plot(x, y)
plt.show()
```

**Başlığı Konumlandır**

Başlığı konumlandırmak için locparametreyi kullanabilirsiniz .title()

Yasal değerler: 'sol', 'sağ' ve 'orta'dır. Varsayılan değer 'orta'dır.

**Örnek
Başlığı sola yerleştirin:**

```python
# BAŞLIK KONUMLANDIRMA : loc parametresi kullanılır.

# ÖRNEK 4: Grafik başlığını sola hizalayalım.
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'left',color='hotpink') # Grafik başlığı sola hizalanmıştır.
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```
