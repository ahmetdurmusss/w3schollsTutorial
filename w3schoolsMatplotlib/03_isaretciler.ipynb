### İŞARETÇİLER 
marker() Her noktayı belirli bir işaretleyici ile vurgulamak için anahtar sözcük bağımsız değişkenini kullanabilirsiniz.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ÖRNEK : Her noktayı bir daire (o) ile işaretleyin.
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
# ÖRNEK : Her noktayı bir daire (*) ile işaretleyin.
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = '*')
plt.show()
```

### FMT
İşaretçiyi belirtmek için kısayol dize notasyonu parametresini de kullanabilirsiniz .

Bu parametreye ayrıca fmt denir ve şu sözdizimiyle yazılır:

marker|line|color

```python
# ÖRNEK : Her noktayı bir daire ile işaretleyin.
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:g')
plt.show()

```

*******************************************************
1) düz çizgi = '-'
2) Noktalı çizgi = ':'
3) Kesik çizgi = '--'
4) Kısa çizgi noktalı çizgie = '-.'
*******************************************************
1) 'r' = kırmızı
2) 'g' = yeşil
3) 'b' = mavi
4) 'c' = cyan
5) 'm' = magenta
6) 'y' = sarı
7) 'k' = siyah
8) 'w' = beyaz

### İŞARETLEYİCİ BOYUTU 
İşaretçilerin boyutunu ayarlamak için anahtar sözcük bağımsız değişkenini "markersize" veya "ms" daha kısa sürümünü kullanabilirsiniz .

```python
# ÖRNEK : İşaretçilerin boyutunu 20 olarak ayarlayın.
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

### İŞARETLEYİCİ RENGİ
İşaretçilerin kenarlarının rengini ayarlamak için anahtar kelime "markeredgecolor" bağımsız değişkenini veya daha kısa olan "mec" kullanabilirsiniz .

```python
# ÖRNEK : KENAR rengini kırmızı olarak ayarlayın.
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

### !!!
İşaretçilerin kenarının içindeki rengi ayarlamak için anahtar kelime " markerfacecolor" bağımsız değişkenini veya daha kısa olan "mfc" kullanabilirsiniz.

```python
# ÖRNEK : YÜZ rengini kırmızı olarak ayarlayın.
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

```python
# ÖRNEK : Hem kenarın hem de yüzün rengini kırmızı olarak ayarlayın.
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r')
plt.show()
```
