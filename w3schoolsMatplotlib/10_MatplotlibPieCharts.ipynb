## PYTHON PASTA GRAFİĞİ
Pyplot ile pasta grafikleri çizmek için pie() fonksiyonunu kullanabilirsiniz.

**Örnek
Basit bir pasta grafiği:**

```python
# ÖRNEK 1 : Pythonda Pie Chart
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 

# Varsayılan olarak ilk bölge çizimi x ekseninden başlar ve saat yönünün tersine hareket eder :kırmızı ve mavi arası yer x ekseni dediğim
```

#### Etiketler 
labels() parametresi ile pasta grafiğine etiketler ekleyin. labels parametresi, her bölge için bir etikete sahip bir dizi olmalıdır.

```python
# ÖRNEK 2 : Etiketli bir pasta grafiği oluşturmak için labels parametresini kullanın.
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

#### Başlangıç ​​Açısı
Belirtildiği gibi varsayılan başlangıç ​​açısı x eksenindedir, ancak bir " startangle " parametre belirleyerek başlangıç ​​açısını değiştirebilirsiniz.

" startanglederece " parametre cinsinden bir açı ile tanımlanır, varsayılan açı 0'dır.

```python
# ÖRNEK 3 : startangle parametresi ile pasta grafiğinin başlangıç açısını belirleyin.
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)#ilk bölgeyi 90 dereceden bşlattık
plt.show() 
```

#### Dilimleme
Bazen bölgelerden birinin öne çıkmasını isteriz, " explode " parametresi bunu yapmamıza olanak tanır. Her değer, her bölgenin merkezden ne kadar uzakta görüntülendiğini temsil eder.

**Örnek
"Elmalar" dilimini turtanın merkezinden 0,2 cm uzağa çekin:**

```python
# ÖRNEK 4 : Explode parametresi ile bir dilim dışarı çıkar.
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"] #dilimlerin isimleri
myexplode = [0.2, 0, 0, 0]  # ilk dilim dışarı çıkar

plt.pie(y, labels = mylabels, explode = myexplode) #dilimlerin isimleri ve ilk dilim dışarı çıkar
plt.show() 
```

#### Pasta Grafiği Gölgelendirme
Pasta grafiğine gölge eklemek için parametreyi shadow=True şeklinde ayarlayın

```python
# ÖRNEK 5 : shadow parametresi ile pasta grafiğine gölge ekleyin.

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

#### Pasta Grafiğini Renklendirme
Her bölgenin rengini " colors " parametresi ile ayarlayabilirsiniz." colors " parametre, belirtilirse, her bölge için bir değere sahip bir dizi olmalıdır.

```python
# ÖRNEK 6 : colors parametresi ile pasta grafiğindeki dilimlerin rengini değiştirin.

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"] 
mycolors = ["black", "hotpink", "b", "#4CAF50"] # dilimlerin renkleri , 4 renk belirledik

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

### Legend
Her bölge için bir açıklama listesi eklemek için şu legend() fonksionunu kullanalım.

```python
# ÖRNEK 7 : legend() fonksiyonu ile pasta grafiğindeki dilimlerin açıklamalarını gösterin.
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

### Başlıklı legend

Efsaneye bir başlık eklemek içi legendın parametresine title  ekleyin.

**Örnek
Başlıklı bir efsane ekleyin:**

```python
# LEGEND() FONKSİYONUNA BAŞLIK EKLEMEK : title parametresi ile başlık ekleyin.

# ÖRNEK 8 : legend() fonksiyonuna başlık ekleyin.
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```
