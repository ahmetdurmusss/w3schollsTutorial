## PYTHON ÇİZGİ GRAFİĞİ 
Çizilen çizginin stilini değiştirmek için " linestyle " veya " ls " daha kısasını kullanabilirsiniz.

1) 'solid' (default)	-----> '-'	
2) 'dotted'	            -----> ':'	
3) 'dashed'	            -----> '--'	
4) 'dashdot'	        -----> '-.'	
5) 'None'	            -----> '' or ' '

```python
# Çizilen çizginin stilini değiştirmek için " linestyle " veya daha kısasını " ls " kullanabilirsiniz :

# ÖRNEK 1 : Çizgiyi noktalı yapmak için "linestyle
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')#,ls='dotted'
plt.show()
```

```python
# ÖRNEK 2 : Çizgiyi kesikli yapmak için "linestyle
plt.plot(ypoints, linestyle = 'dashed')
```

```python
# ÖRNEK 3 : Çizgiyi noktalı ve kesikli yapmak için "linestyle = 'dotted-dashed' veya "linestyle = ':' yazarız. 
plt.plot(ypoints, ls = ':')
```

#### Çizgi Rengi
Satırın rengini ayarlamak için " color " veya daha kısa olan " c "  kullanabilirsiniz.

```python
# ÖRNEK 4 : Çizgi rengini değiştirmek için "color" parametresini kullanabiliriz. 

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'y') # yellow
plt.show()
```

```python
# ÖRNEK 5 : Çizgi rengini değiştirmek için "color" parametresini kullanabiliriz. Parametreye bir renk kodu da verebiliriz.
plt.plot(ypoints, c = '#4CAF50') # RGB Hex code
```

Örnek
"Sıcak pembe" adlı renge ait arsa:

```python
# ÖRNEK 6 : Çizgi rengi için renk adı kullanabiliriz.
plt.plot(ypoints, c = 'hotpink') # CSS color names
```

#### ÇİZGİ KALINLIĞI
Satırın genişliğini değiştirmek için " linewidth " veya daha kısa olan " lw "  kullanabilirsiniz.

```python
# ÖRNEK 7 : Çizgi genişliğini değiştirmek için "linewidth" parametresini kullanabiliriz.
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20') # 20px genişliğinde çizgi
plt.show()
```

#### Aynı Tuvalde Birden Fazla Grafik Çizimi

" plt.plot() " fonksiyonu ile istediğiniz kadar çizgi çizebilirsiniz .

Örnek
plt.plot()Her çizgi için bir fonksiyon belirleyerek iki çizgi çizin :

```python
# ÖRNEK 8 : İki çizgi çizmek için iki plot() fonksiyonu kullanabiliriz.

import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10]) # 1. çizgi
y2 = np.array([6, 2, 7, 11]) # 2. çizgi

plt.plot(y1,c='hotpink') # 1. çizgi
plt.plot(y2) # 2. çizgi

plt.show()
```

```python
# ÖRNEK 9 : İki çizgiyi aynı eksenlerde çizmek için tek bir plot() fonksiyonu kullanabiliriz.

import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3]) # 1. çizginin x eksenindeki değerleri
y1 = np.array([3, 8, 1, 10])  # 1. çizginin y eksenindeki değerleri
x2 = np.array([0, 1, 2, 3]) # 2. çizginin x eksenindeki değerleri
y2 = np.array([6, 2, 7, 11]) # 2. çizginin y eksenindeki değerleri

plt.plot(x1, y1, x2, y2) # 2 çizgi
plt.show() 
```
